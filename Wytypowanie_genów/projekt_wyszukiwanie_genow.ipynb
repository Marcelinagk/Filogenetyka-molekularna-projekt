```python
import pandas as pd

# Wczytanie plików
mrsa = pd.read_csv("MRSA.csv")
mssa = pd.read_csv("MSSA.csv")
sepi = pd.read_csv("SEPI.csv")

# Zbiory genów z kolumny 'Genes'
genes_mrsa = set(mrsa["Genes"].dropna())
genes_mssa = set(mssa["Genes"].dropna())
genes_sepi = set(sepi["Genes"].dropna())
print(f"Liczba genów MRSA: {len(genes_mrsa)}")
print(f"Liczba genów MSSA: {len(genes_mssa)}")
print(f"Liczba genów SEPI: {len(genes_sepi)}")
```

```python
# Zbiory genów z kolumny 'PGFam'
genes_mrsa = set(mrsa["PGFam"].dropna())
genes_mssa = set(mssa["PGFam"].dropna())
genes_sepi = set(sepi["PGFam"].dropna())
print(f"Liczba genów MRSA: {len(genes_mrsa)}")
print(f"Liczba genów MSSA: {len(genes_mssa)}")
print(f"Liczba genów SEPI: {len(genes_sepi)}")

# Usunięcie genów wspólnych dla MRSA i MSSA
common_mrsa_mssa = genes_mrsa.difference(genes_mssa)

# Usunięcie genów obecnych w SEPI
result_genes = common_mrsa_mssa.difference(genes_sepi)


# Wypisanie wyników
for gene in sorted(result_genes):
    print(gene)

print(f"\nLiczba genów: {len(result_genes)}")
```

```python
# Znalezienie genów (kolumna Genes) odpowiadających nazwom PGFam w result_genes
subset = mrsa[mrsa["PGFam"].isin(result_genes)]

# geny (kolumna Genes) odpowiadające tym PGFam
genes_for_targets = subset["Genes"].unique()

print("Znalezione geny (kolumna Genes) dla podanych nazw PGFam:")
for g in genes_for_targets:
    print(g)
```
